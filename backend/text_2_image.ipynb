```python
from quickdraw import QuickDrawData

qd = QuickDrawData()
print(len(qd.drawing_names))
```

```python
import spacy_universal_sentence_encoder
# load one of the models: ['en_use_md', 'en_use_lg', 'xx_use_md', 'xx_use_lg']
nlp = spacy_universal_sentence_encoder.load_model('en_use_lg')
# get two documents
doc_1 = nlp('Hi there, how are you?')
doc_2 = nlp('dog')
# use the similarity method that is based on the vectors, on Doc, Span or Token
print(doc_1.similarity(doc_2))
```

```python
score = {}
for a in list(qd.drawing_names):
    doc_2 = nlp(a)
    score[a] = doc_1.similarity(doc_2)
cat = sorted(score)
doodle_cat = cat[0]
print(doodle_cat)
```

```python
from quickdraw import QuickDrawDataGroup

ants = QuickDrawDataGroup(doodle_cat)
ant = ants.get_drawing()
ant.image.save("/home/cnarte/hackathons/APril/backend/doodlesmy_ant.gif")
```
